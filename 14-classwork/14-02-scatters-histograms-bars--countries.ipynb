```python
import pandas as pd
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv('sources/countries.csv')
df.head()
```

```python
bins = range(0,101,5)
df.life_expectancy.hist()
```

```python
bins = range(0,101,5)
df.life_expectancy.hist(bins=bins, figsize=(18,5))
```

```python
bins = range(0,101,5)
ax = df.life_expectancy.hist(bins=bins, figsize=(18,5))
ax.set_xlim(30,100)
ax.set_title("This is a histogram")
ax.grid(False)
ax.axhline(30)
```

```python
bins = range(0,101,5)
ax = df.life_expectancy.hist(bins=bins, figsize=(18,5), color='lightblue')
ax.set_xlim(30,100)
ax.set_title("This is a histogram", color='red')
ax.set_yticks([10,15,20,25,30, 40,50,60])

plt.grid(False)
plt.axhline(30, color='red')
```

```python
ax = df.sort_values(by='life_expectancy').plot(x='Country', y='life_expectancy', kind='bar', width=1, figsize=(18,5))
ax.set_xticks([])
```

```python
# Building colors to pass to matplotlib or Seaborn
# Use as color=colors for matplotlib, palette=colors for Seaborn
def build_colors(row):
    if row['Country'] == 'Switzerland':
        return 'red'
    elif row['Country'] == 'Germany':
        return 'red'
    else:
        return 'lightgrey'

colors = df.reset_index().apply(build_colors, axis=1)

ax = df.sort_values(by='life_expectancy') \
    .plot(x='Country', y='life_expectancy', kind='bar', width=1, figsize=(18,5), color=colors)

ax.set_xticks([])
```

```python
df['Pop_reduce'] = df.Population / 1000000
```

```python
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', figsize=(18,5), color=colors)
```

```python
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', figsize=(18,5), s=df.Pop_reduce, alpha=0.6)
```

```python
sns.distplot(df.life_expectancy)
```

```python
bins=range(0,101,5)
sns.distplot(df.life_expectancy, kde=False, bins=bins)
```

```python
bins=range(0,101,5)
ax = sns.distplot(df.life_expectancy, kde=False, bins=bins)
ax.set_xlim(30,100)
ax.set_title("This is a histogram")
ax.grid(False)
ax.axvline(df.life_expectancy.median(), color='red')
```
