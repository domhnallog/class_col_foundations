```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv("sources/gender.csv")
df.head()
```

```python
df.plot(x='state', y='male', kind='barh', figsize=(18,5))
```

```python
df.plot(x='state', y='male', kind='barh', color='teal', figsize=(18,5))
```

```python
colors = ['lightgrey','teal','lightgrey','lightgrey']
df.plot(x='state', y='male', kind='barh', color=colors, figsize=(18,5))
```

```python
colors = ['lightblue','pink']
df.plot(x='state', y=['male','female'], kind='barh', color=colors, figsize=(18,5))
```

```python
colors = ['lightblue','pink']
df.plot(x='state', y=['male','female'], kind='barh', color=colors, figsize=(18,5), stacked=True)
```

```python
df.plot(x='state', y='male', kind='barh', color='gold', figsize=(18,5), width=0.1)
```

```python
df['pct_male'] = df.male / (df.male + df.female)
df['pct_female'] = df.female / (df.male + df.female)
df
```

```python
df.plot(x='state', y=['pct_male', 'pct_female'], kind='barh', figsize=(18,5), stacked=True)
plt.tight_layout()
plt.savefig('output.png')
```

```python
df.plot(x='state', y=['pct_male', 'pct_female'], kind='barh', figsize=(18,5), stacked=True)
plt.tight_layout()
plt.savefig('output.pdf')
```

```python
df.plot(x='state', 
        y=['pct_male', 'pct_female'], 
        kind='barh', 
        figsize=(18,5), 
        stacked=True,
        title ='!!!STUFF!!!'
       )
```
