```python
df = pd.read_csv("sources/countries.csv")
df.head()
```

```python
# Building colors to pass to matplotlib or Seaborn
# Use as color=colors for matplotlib, palette=colors for Seaborn
def build_colors(row):
    if row['Country'] == 'Switzerland':
        return 'green'
    elif row['Country'] == 'Germany':
        return 'blue'
    else:
        return 'lightgrey'

colors = df.reset_index().apply(build_colors, axis=1)
colors.head()
```

```python
colors.dtypes
```

```python
ax = df.sort_values(by='life_expectancy') \
    .plot(x='Country', y='life_expectancy', kind='bar', width=1, color=colors)

ax.set_xticks([])
```

# Scatterplot

```python
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter')
```

```python
# Building colors to pass to matplotlib or Seaborn
# Use as color=colors for matplotlib, palette=colors for Seaborn
def build_colors(row):
    if row['Country'] == 'France':
        return 'red'
    elif row['Country'] == 'India':
        return 'red'
    else:
        return 'lightgrey'

colors = df.reset_index().apply(build_colors, axis=1)

ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', color=colors)
```

```python
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', s=50)
```

```python
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', s=[10,100])
```

```python
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', s=df.Population)
```

```python
sizes = df.Population.head() / 100000
ax = df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter', s=sizes, alpha=0.5)
```
