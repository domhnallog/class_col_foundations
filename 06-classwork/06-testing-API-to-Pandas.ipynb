```python
import pandas as pd
import requests

%matplotlib inline
```

```python
from dotenv import load_dotenv
load_dotenv()
import os
secretapikey = os.getenv("darksky_apikey")
base_url = "https://api.darksky.net/forecast/"
```

```python
location = ("37.8267"+","+"122.4233")
response = requests.get(f'{base_url}{secretapikey}/{location}')
data = response.json()
degree_sign = u'\N{DEGREE SIGN}'
```

```python
# Check hourly
data['hourly']
```

```python
df = pd.DataFrame(data['hourly']['data'])
```

```python
#  NaN = missing data / empty field
#  Also called NA
#  As in dropna
#  If you set it to 'false' it will count those empty values
df.precipType.value_counts(dropna=False)
```

```python
#  You can do math over the whole list
df.precipType.value_counts(dropna=False) / 49
```

```python
#  But as the total number may vary from list to list, you can auto set the reslts to percentages with normalize
df.precipType.value_counts(dropna=False, normalize=True)
```
