# Putting things in groups/categories

```python
# Just guess to figure out what goes where, more or less
df.groupby(by='Continent').GDP_per_capita.median()
```

```python
df.groupby(by='Continent').GDP_per_capita.median().sort_values()
```

```python
# The backslash means "pretend I didn't hit enter"
df.groupby(by='Continent') \
    .GDP_per_capita.median() \
    .sort_values(ascending=False)
```

```python
# Give me the three continents with the most people on them
# aka, add up the people on each continent, and give me the top 3
df.groupby(by='Continent') \
    .Population \
    .sum() \
    .sort_values(ascending=False) \
    .head(3)
```

```python
# ORDER BY desc
df.groupby(by='Continent') \
    .GDP_per_capita.median().sort_values(ascending=False)
```

# Calculating things!

```python
# We have GDP per capita
# We have the population
# Calculate the total GDP of each country
# Math tip: You calculate the total GDP by multiplying
# the population by the GDP per capita
df.Population * df.GDP_per_capita
```

```python
# If you want to make a new column, you have to do
# the quotation marks dictionary-style thing or it won't work
df['Total_GDP'] = df.Population * df.GDP_per_capita
```

```python
# df.Continent = 'Asia'
```

```python
# True on every row where it's true
# False on every row where it isn't true
df.Continent == 'Asia'
```

```python
# To filter rows, you put this inside of df[]
df[df.Continent == 'Asia']
```

```python
# Get the total population of every country in Asia
df[df.Continent == 'Asia'].Population.sum()
```

```python
# Get the total population of Asia in MILLIONS
df[df.Continent == 'Asia'].Population.sum() / 1000000
```

```python
# If you have multiple conditionals/clauses
# in pandas, you have to use parens and &
df[(df.Continent == 'Asia') & (df.Population < 5000000)]
```

```python
df[ \
   (df.Continent == 'Asia') \
   &
   (df.Population < 5000000) \
]
```

```python
is_in_asia = df.Continent == 'Asia'
is_in_na = df.Continent == 'N. America'
smaller_than_5m = df.Population < 5000000
df[is_in_asia & smaller_than_5m]
```

```python
# "or" is |
df[(is_in_asia | is_in_na) & smaller_than_5m]
```

```python
df[(df.Continent == 'Asia') & (df.Population < 5000000)].sort_values(by='Population')
```

```python
# Step 1. Filter
# Step 2. Groupby 
# Step 3. Select columns
# Step 4. Math

# The median population of countries
# with a life expectancy under 75 years
# on each continent
```

```python
df[df.life_expectancy < 75].groupby(by='Continent').Population.median()
```
