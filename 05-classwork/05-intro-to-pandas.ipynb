## Adding Data, Combining Data & Defining Variables

```python
# This provides just a list of whether true or false
df.Continent == 'Asia'
```

```python
# This lists just those Asia countries
df[df.Continent == 'Asia']
```

```python
# Get total pop of Asia:
df[df.Continent == 'Asia'].Population.sum()
```

```python
# Get total pop of Asia in millions:
df[df.Continent == 'Asia'].Population.sum() / 1000000
```

```python
# Two variables:
# Don't use sample below it tries to use python language in pandas
#    AND is  &
#    OR  is  |  (pipe)
# df[df.Continent == 'Asia' and df.Population < 5000000]
```

```python
# Joining tables on two variables:
df[(df.Continent == 'Asia') & (df.Population < 5000000)]
```

```python
# Joining tables on two variables, sorted:
df[(df.Continent == 'Asia') & (df.Population < 5000000)].sort_values(by='Population', ascending=False)
```

```python
# Defining variables
ctrys_asia = df.Continent == 'Asia'
ctrys_na = df.Continent == 'N. America'
pop_under5m = df.Population < 5000000

ctrys_asia.describe()
```

```python
pop_under5m.describe()
```

```python
# Joining defined variables, 'AND' is '&'
df[ctrys_asia & pop_under5m]
```

```python
# Joining defined variables, 'OR' is '|'
df[ctrys_asia | ctrys_na & pop_under5m]
```

```python
# Find the median population for those countries with a life exp below 75 on each continent
#     Filter 
#     Group only for continent / select columns
#     Process median math
#     Arrange list by numbers and list them largest to smallest 

df[df.life_expectancy < 75].groupby(by='Continent').Population.median().sort_values(ascending=False)
```
