```python
import requests
import re
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.bbc.com/')
doc = BeautifulSoup(response.text)
```

```python
titles = doc.find_all(class_='media__title')
for title in titles:
    print(title.text.strip())
```

```python
links = doc.find_all(class_='media__link')
for link in links:
    print(link['href'])
```

```python
summaries = doc.find_all(class_='media__summary')
for summary in summaries:
    print(summary.text.strip())
```

```python
stories = doc.find_all(class_='media-list__item')
stories
```

```python
for story in stories:
    print("-----------------------------------------------------------------------------------------")
    title = story.find(class_='media__title').text.strip()
    link = story.find(class_='media__link')['href']
    summary = story.find(class_='media__summary').text.strip()
    print(title, '--', link, '--', summary)
```

```python
for story in stories:
    print("-----------------------------------------------------------------------------------------")
    try:
        title = story.find(class_='media__title').text.strip()
    except:
        title = 'Title not found'
    try:
        link = story.find(class_='media__link')['href']
    except:
        link = 'Link not found'
    try:
        summary = story.find(class_='media__summary').text.strip()
    except:
        summary = 'Summary not found'
    print(title, '--', link, '--', summary)
```
