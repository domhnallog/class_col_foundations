```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.congress.gov/members')
doc = BeautifulSoup(response.text)
```

```python
### TEST VIEWS OF DATA
# print(doc)
print(doc.prettify())
```

```python
members = doc.find_all(class_='expanded')
for member in members:
    print(member.prettify())
    print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
```

```python
rows = []
for member in members:
    row = {}
    row['link'] = member.a['href']
    row['name_raw'] = member.a.text
    rows.append(row)
#     print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
rows
df = pd.DataFrame(rows)
df.head()
```

```python
# see if page indexing changes url
# eg https://www.congress.gov/members?pageSize=250&page=10
```

```python
# give it a range 1 bigger than the last know page (so 11 for known page 10)
for page_num in range(1,11):
    url = f"https://www.congress.gov/members?pageSize=250&page={page_num}"
    print(url)
```

```python
rows = []
for page_num in range(1,11):
    url = f"https://www.congress.gov/members?pageSize=250&page={page_num}"
    print(f"Now scraping {url}")
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    members = doc.find_all(class_='expanded')
    for member in members:
        row = {}
        row['link'] = member.a['href']
        row['name_raw'] = member.a.text
        rows.append(row)
rows
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.shape
```

```python
df['slug'] = df.link.str.replace('https://www.congress.gov/member/','')
df.head()
```

```python
df.to_csv('congress.csv', index=False)
```
