```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.bbc.com/')
doc = BeautifulSoup(response.text)
```

# Step 1: Find the tags or selectors for the data you're interested in

```python
titles = doc.find_all(class_='media__title')
for title in titles:
    print(title.text.strip())
```

```python
links = doc.find_all(class_='media__link')
for link in links:
    print(link['href'])
```

```python
summaries = doc.find_all(class_='media__summary')
for summary in summaries:
    print(summary.text.strip())
```

```python
rows = []

# Find every full story
stories = doc.find_all(class_='media-list__item')

# And loop through them
for story in stories:
    row = {}

    # print("-----------")
    # Find the title of that story
    try:
        row['title'] = story.find(class_='media__title').text.strip()
    except:
        pass

    try:
        row['url'] = story.find(class_='media__link')['href']
    except:
        pass
    
    try:
        row['summary'] = story.find(class_='media__summary').text.strip()
    except:
        # Just don't do anything
        pass
        # print("Failed to find summary")

    # Take our dictionary we just made
    # and add it to our list of dictionaries
    rows.append(row)
    # print(row)

df = pd.DataFrame(rows)
df
```
