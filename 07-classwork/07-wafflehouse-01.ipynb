```python
pop_df = pd.read_excel("wafflehouses/nst-est2016-01.xlsx", skiprows=3, skipfooter=7)
```

```python
pop_df.head()
```

```python
pop_df = pop_df.rename(columns={
    "Unnamed: 0": "State"
})
pop_df.head(10)
```

```python
pop_df['State'] = pop_df.State.str.replace(".","")
pop_df.head(10)
```

```python
abbr_df = pd.read_excel("wafflehouses/abbreviations.xlsx")
```

```python
abbr_df.head(10)
```

```python
df.merge(abbr_df, left_on='state_abbr', right_on='abbreviation')
```

```python
merged = df.merge(abbr_df, left_on='state_abbr', right_on='abbreviation')
merged.head()
```

```python
merged.merge(pop_df, left_on='name', right_on='State')
```

```python
merged_pop = merged.merge(pop_df, left_on='name', right_on='State')
merged_pop.head()
```

```python
# It made a series, but you turn it into a dataframe
count_df = merged_pop.State.value_counts().to_frame()
count_df.head()
```

```python
# count_df doesn't have a column name for the state, it just has an index
calc_df = count_df.merge(pop_df, right_on='State', left_index=True)
calc_df.head()
```
