```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dllr.state.md.us/cgi-bin/ElectronicLicensing/OP_Search/OP_search.cgi'
response = requests.get(url, verify=False)
doc = BeautifulSoup(response.text)
```

```python
rows = doc.find_all('tr')
for row in rows:
    print(row.text)
```
